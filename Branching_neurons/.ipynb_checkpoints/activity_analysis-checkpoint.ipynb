```python
# Import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Load data

```python
# Load activity files
SIZE = 80
sigma_vals = [0.7, 0.8, 0.9, 1, 1.1, 1.2, 1.5, 1.8, 2, 2.5]
file_paths = ['activity_sigma_'+str(n)+'.txt' for n in sigma_vals]
sigma_dfs = [pd.read_csv(f, comment='#', sep=' ', index_col='Step') for f in file_paths]
mean_activations = [s.mean() / (SIZE ** 2) for s in sigma_dfs]
```

```python
FONTSIZE = 25
fig, ax = plt.subplots(figsize=(13,8))
plt.plot(sigma_vals,mean_activations, lw=5)
plt.xticks(fontsize=FONTSIZE-2)
plt.yticks(fontsize=FONTSIZE-2)
plt.xlabel(r'$\sigma$', fontsize=FONTSIZE)
plt.ylabel(r'$\rho$', fontsize=FONTSIZE)
plt.xlim(0.7,2.5)
plt.ylim(-0.02, 1)
plt.grid(True);
```

# Critical point

```python
# Load critical point data
# 1st file
df_critical = pd.read_csv('critical_activity_sigma_1_steps1e4.txt', 
                          comment='#', sep=' ', index_col='Step')
# Additional files
additional_files = ['critical_activity_sigma_1_steps3e4.txt','critical_activity_sigma_1_steps2e4.txt',
                    'critical_activity_sigma_1_steps2e4_2.txt','critical_activity_sigma_1_steps2e4_3.txt']
for f in additional_files:
    df_critical_temp = pd.read_csv(f, comment='#', sep=' ', index_col='Step')
    df_critical_temp.index = df_critical_temp.index + df_critical.index.max()
    df_critical = pd.concat([df_critical,df_critical_temp], axis=0)

# Get avalanches info
all_avalanches = {'Size':[], 'Start':[], 'End':[], 'Length':[]}
starting_points = []
ending_points = []
# Get starting and ending points for all avalanches
for i in range(df_critical.shape[0]-1):
    prev_step = df_critical.iloc[i]['activity']
    next_step = df_critical.iloc[i+1]['activity']
    if prev_step == 0 and next_step > 0:
        starting_points.append(i+1)
    elif (prev_step > 0 and next_step == 0) or i == df_critical.shape[0]-2:
        ending_points.append(i)

# Get avalanches size, start, end and duration
for i in range(len(starting_points)):
    start = starting_points[i]
    end = ending_points[i]
    size = df_critical.iloc[start:end+1]['activity'].sum()
    # Save avalanche to dict
    all_avalanches['Size'].append(size)
    all_avalanches['Start'].append(start)
    all_avalanches['End'].append(end)
    all_avalanches['Length'].append((end+1)-start)
all_avalanches = pd.DataFrame(all_avalanches)
    
# Now let's bin these avalanches to get and estimate 
# of their probability
RELATIVE = 0.001
TOP = 10
L_RELATIVE = 0.001
L_TOP = 1
binned_avalanches = {'Size': [], 'Count': []}
binned_lengths = {'Length': [], 'Count': [], 'Avg_size': []}
for i in range(all_avalanches.shape[0]):
    # Bin by size
    size = all_avalanches.iloc[i]['Size']
    delta_size = int(round(size * RELATIVE, 0))
    if delta_size > TOP:
        delta_size = TOP
    size_range = [size + n for n in range(-delta_size, delta_size+1)]
    check =  any(item in binned_avalanches['Size'] for item in size_range)
    if check:
        # Find matching value
        for s in size_range:
            if s in binned_avalanches['Size']:
                binned_avalanches['Count'][
                    binned_avalanches['Size'].index(s)] += 1
    else:
        binned_avalanches['Size'].append(size)
        binned_avalanches['Count'].append(1)
    
    # Bin by length
    length = all_avalanches.iloc[i]['Length']
    delta_length = int(round(length * L_RELATIVE, 0))
    if delta_length > L_TOP:
        delta_length = L_TOP
    length_range = [length + n for n in range(-delta_length, delta_length+1)]
    check_length = any(item in binned_lengths['Length'] for item in length_range)
    if check_length:
        # Find matching value
        for l in length_range:
            if l in binned_lengths['Length']:
                binned_lengths['Count'][
                    binned_lengths['Length'].index(l)] += 1
    else:
        binned_lengths['Length'].append(length)
        binned_lengths['Count'].append(1)
        
        # Get average avalanche size for a given duration
        avg_size = all_avalanches.loc[all_avalanches['Length'] == length, 
                                      'Size'].mean()
        binned_lengths['Avg_size'].append(avg_size)

# Get prob for binned sizes
binned_avalanches = pd.DataFrame(binned_avalanches)
binned_avalanches['Prob'] = binned_avalanches['Count'] / binned_avalanches['Count'].sum()

# Get prob for binned lengths
binned_lengths = pd.DataFrame(binned_lengths)
binned_lengths['Prob'] = binned_lengths['Count'] / binned_lengths['Count'].sum()
    
```

```python
# Calculate critical exponent for avalanche size
reg_data = binned_avalanches.loc[binned_avalanches['Size'] < 80]
X = np.array(np.log10(reg_data['Size'])).reshape(-1, 1)
y = np.log10(reg_data['Prob'])
size_linear = LinearRegression().fit(X,y)
size_linear_data = [
    size_linear.coef_[0] * x +\
    size_linear.intercept_ for x in np.log10(np.sort(reg_data['Size']))]

# Save parameter
tau = size_linear.coef_[0]
theory_tau = -1.46

FONTSIZE = 25
fig, ax = plt.subplots(figsize=(13,8))
plt.scatter(np.log10(binned_avalanches['Size']), 
            np.log10(binned_avalanches['Prob']), s=70, c='b',
            label='Simulation')
plt.plot(np.log10(np.sort(reg_data['Size'])),
         size_linear_data, c='k', lw=2, ls='--',
         label=r'Model, $\tau = %.2f$'%size_linear.coef_[0])
plt.plot(np.log10(np.sort(reg_data['Size'])),
         np.log10(np.sort(reg_data['Size'])) * theory_tau +\
         np.log10(binned_avalanches['Prob'].max()), 
         c='g', lw=2, ls='--',
         label=r'Theory, $\tau = %.2f$'%theory_tau)
plt.xlabel(r'$\log(S)$', fontsize=FONTSIZE)
plt.ylabel(r'$\log(P(S))$', fontsize=FONTSIZE)
plt.xticks(fontsize=FONTSIZE-2)
plt.yticks(fontsize=FONTSIZE-2)
plt.legend(fontsize=FONTSIZE)
plt.grid(True);
```

```python
# Calculate critical exponent for avalanche duration
reg_data = binned_lengths.loc[
    (binned_lengths['Length'] < 80) & (binned_lengths['Length'] > 1)]
X = np.array(np.log10(reg_data['Length'])).reshape(-1, 1)
y = np.log10(reg_data['Prob'])
length_linear = LinearRegression().fit(X,y)
length_linear_data = [
    length_linear.coef_[0] * x +\
    length_linear.intercept_ for x in np.log10(np.sort(reg_data['Length']))]

# Save parameter
alpha = length_linear.coef_[0]
theory_alpha = -1.61

FONTSIZE = 25
fig, ax = plt.subplots(figsize=(13,8))
plt.scatter(np.log10(binned_lengths['Length']), 
            np.log10(binned_lengths['Prob']), s=70, c='b',
            label='Simulation')
plt.plot(np.log10(np.sort(reg_data['Length'])),
         length_linear_data,  c='k', lw=2, ls='--',
         label=r'Model, $\alpha = %.2f$'%length_linear.coef_[0])
plt.plot(np.log10(np.sort(reg_data['Length'])),
         np.log10(np.sort(reg_data['Length'])) * theory_alpha +\
         length_linear.intercept_, 
         c='g', lw=2, ls='--',
         label=r'Theory, $\alpha = %.2f$'%theory_alpha)
plt.xlabel(r'$\log(T)$', fontsize=FONTSIZE)
plt.ylabel(r'$\log(P(T))$', fontsize=FONTSIZE)
plt.xticks(fontsize=FONTSIZE-2)
plt.yticks(fontsize=FONTSIZE-2)
plt.legend(fontsize=FONTSIZE)
plt.grid(True);
```

```python
# Calculate gamma parameter
gamma = (alpha - 1) / (tau - 1)
theory_gamma = (theory_alpha - 1) / (theory_tau - 1)
# Get linear model
reg_data = binned_lengths.loc[
    (binned_lengths['Length'] <= 50) & (binned_lengths['Avg_size'] <= 50)]
length_size_linear_data = [
    gamma * x for x in np.log10(np.sort(reg_data['Length']))]



FONTSIZE = 25
fig, ax = plt.subplots(figsize=(13,8))
plt.scatter(np.log10(binned_lengths['Length']), 
            np.log10(binned_lengths['Avg_size']), s=70, c='b',
            label='Simulation')
plt.plot(np.log10(np.sort(reg_data['Length'])),
         length_size_linear_data,  c='k', lw=2, ls='--',
         label=r'Model, $\gamma = %.2f$'%gamma)
plt.plot(np.log10(np.sort(reg_data['Length'])),
         np.log10(np.sort(reg_data['Length'])) * theory_gamma, 
         c='g', lw=2, ls='--',
         label=r'Theory, $\gamma = %.2f$'%theory_gamma)
plt.xlabel(r'$\log(T)$', fontsize=FONTSIZE)
plt.ylabel(r'$\log(\bar{S})$', fontsize=FONTSIZE)
plt.xticks(fontsize=FONTSIZE-2)
plt.yticks(fontsize=FONTSIZE-2)
plt.legend(fontsize=FONTSIZE)
plt.grid(True);
```

```python
# Let's visualize the time profile of some avalanches
FONTSIZE = 25
LW = 2
ALPHA = 0.025
fig, (ax1, ax2) = plt.subplots(figsize=(26,8), nrows=1, ncols=2)
AV_ID = [n for n in range(len(starting_points))]
avalanche_examples = []
for i in AV_ID:
    avalanche_case = df_critical.iloc[
        starting_points[i]:ending_points[i]+1].copy()
    if avalanche_case.shape[0] > 20:
        avalanche_case['Rel_step'] = avalanche_case.index - avalanche_case.index.min() + 1
        avalanche_case['Norm_step'] = (avalanche_case['Rel_step'] - avalanche_case['Rel_step'].min()) /\
                                        avalanche_case['Rel_step'].max()
        avalanche_case['Scaled_activity'] = (avalanche_case['activity'] -\
                                             avalanche_case['activity'].min()) / avalanche_case['activity'].max()
        
        avalanche_examples.append(avalanche_case)
        ax1.plot(np.log10(avalanche_case['Rel_step']), avalanche_case['activity'], 
                 lw=LW, c='k', alpha=ALPHA)
        ax2.plot(avalanche_case['Norm_step'], avalanche_case['Scaled_activity'], 
                 lw=LW, c='k', alpha=ALPHA)
        
for ax in [ax1,ax2]:
    ax.tick_params(labelsize=FONTSIZE-2)
ax1.set_xlabel(r'$\log(t)$', fontsize=FONTSIZE)   
ax1.set_ylabel(r'$N_{neurons}$', fontsize=FONTSIZE)
ax2.set_xlabel(r'$t/T$', fontsize=FONTSIZE)
ax2.set_ylabel(r'$\tilde{N}_{neurons}$', fontsize=FONTSIZE);
```
